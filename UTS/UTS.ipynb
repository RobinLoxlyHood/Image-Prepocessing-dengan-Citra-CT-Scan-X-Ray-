# Import Library

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

# Baca Image

```python
img1 = cv.imread('covid_image.png') #membaca image pertama
img2 = cv.imread('mask.png') #membaca image kedua
```

```python
img1_gray = cv.cvtColor(img1,cv.COLOR_BGR2GRAY) #ubah gmabar1 ke grayscale
img2_gray = cv.cvtColor(img2,cv.COLOR_BGR2GRAY) #ubah gmabar2 ke grayscale
```

```python
img1_gray.shape #mengecek layer grayscale
```

```python
img2_gray.shape #mengecek layer grayscale
```

# Operasi Bitvise AND

```python
hasil=cv.bitwise_and(img1_gray,img1_gray,mask = img2_gray) #Melakukan operasi bitvise and untuk menggabungkan kedua gambar
```

```python
cv.imshow('Cek hasil Bitvise', hasil) #Menampilkan hasil untuk mengecek apakah hasil operasi bitvise sudah sesuai
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imwrite('res_bitwise.jpg',hasil) # Membuat citra output hasil Operasi bitvise
```

# Noise Removal dengan Filter Spasial

```python
blur_average = cv.blur(hasil,(3,3)) #Melakukan noise removal dengan average kernel 3x3
blur_gaussian = cv.GaussianBlur(hasil,(3,3),0) #Melakukan noise removal dengan GaussianBlur kernel 3x3
blur_median = cv.medianBlur(hasil,3) #Melakukan noise removal dengan medianBlur kernel 3x3
```

```python
cv.imshow('Smoothing Average Filter', blur_average) #menampilkan Hasil noise removal dengan average
cv.imshow('Smoothing Gaussian', blur_gaussian) #menampilkan Hasil noise removal dengan GaussianBlur
cv.imshow('Smoothing Median Filter', blur_median) #menampilkan Hasil noise removal dengan medianBlur
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imwrite('res_noise_removal.jpg',blur_median) # Membuat citra output hasil noise removal dengan medianBlur
```

# Image Histogram

```python
hist,bins = np.histogram(blur_median.flatten(),256,[0,256]) #membuat hist dan bins
cdf = hist.cumsum() #cdf adalah sum dari nilai hist
cdf_m = np.ma.masked_equal(cdf,0) #cdf_m adalah masking nilai 0/ menghilangkan nilai 0
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min()) # perjumlahan untuk mencari sk*255
cdf = np.ma.filled(cdf_m,0).astype('uint8') #round atau pembulatan nilai terdekat menghasilkan nilai histogram yang baru
img2 = cdf[blur_median] # nilai histogram yang baru implementasi ke img 
cdf_normalized = cdf * float(hist.max()) / cdf.max() #melakukan normalisasi nilai img histogram (menggantikan nilai image histogram yang lama ke yang baru)
```

```python
# Menampilkan Plot Histogram sebelum dilakukan HistEqulizer
plt.subplot(2, 1, 1)
plt.plot(cdf, color = 'b')
plt.hist(blur_median.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.savefig('hist_awal.jpg')
plt.show()
```

```python
# Menampilkan Plot Histogram sesudah dilakukan HistEqulizer
plt.subplot(2, 1, 2)
plt.plot(cdf_normalized, color = 'b')
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.savefig('hist_akhir.jpg')
plt.show()
```

```python
cv.imshow('img1',blur_median) #menampilkan image sebelum dilakukan HistEqulizer
cv.imshow('img2',img2) #menampilkan image sesudah dilakukan HistEqulizer
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imwrite('res_contras.jpg',img2) # Membuat citra output hasil HistEqulizer
```

# Thresholding

```python
ret,thresh1 = cv.threshold(img2,127,255,cv.THRESH_BINARY) #melakuakan thresholding binary dengan parameter min 127 max 255
ret,thresh2 = cv.threshold(img2,127,255,cv.THRESH_BINARY_INV) #melakuakan thresholding binary_inv dengan parameter min 127 max 255
ret,thresh3 = cv.threshold(img2,127,255,cv.THRESH_TRUNC)  #melakuakan thresholding TRUNC dengan parameter min 127 max 255
ret,thresh4 = cv.threshold(img2,127,255,cv.THRESH_TOZERO) #melakuakan thresholding TOZERO dengan parameter min 127 max 255
ret,thresh5 = cv.threshold(img2,127,255,cv.THRESH_TOZERO_INV) #melakuakan thresholding TOZERO_INV dengan parameter min 127 max 255
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV'] #Title
images = [img2, thresh1, thresh2, thresh3, thresh4, thresh5] #image
#Melakukan looping untuk menampilkan hasil thresholding
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
